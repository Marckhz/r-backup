# 2. ¿Cuál es el grupo de edad con el mayor número total de colisiones reportadas? Puede considerar cualquier grupo de edad a los fines de esta pregunta. Explique esto de manera gráfica.

```python
age_series = dataframe['V1_Edad']
```

```python
age_series = age_series.fillna(0)
```

```python
age_series = age_series.replace(to_replace='SD', value=0)
age_series
```

```python
labels=['18-30','30-40', '40-50', '50-40','60-70','70-80','80-90']
age_groups = pd.cut(age_series.astype(int), bins=[18,30,40,50,60,70,80,90], 
                            labels=labels).value_counts()
```

```python
age_groups
```

```python
plt.figure(figsize=(12,8))
ax = age_groups[age_groups > 0].sort_values(ascending=True).plot(kind='barh', 
                 fontsize=14)
ax.set_title('Numero de reportes de Colision agrupado por Edades',fontsize=18)
ax.set_xlabel('Frecuencia de reportes', fontsize=14)
ax.set_ylabel('Grupo de Edades', fontsize=14)
```

### Mostrando mediante un grafico la frencuencia de reportes divido en grupos de Edades. Podemos Observar que el grupo de edad que mas reportes de colisiones notifica esta en el intervalo de 30-40 años.

## 3. Si tuviera que ver una colisión en el camino "Garza Sada", ¿cuáles serían las probabilidades de que esa persona tuviera una prueba positiva sobre la ingestión de alcohol?


```python
index_where_zero= street_alcohol_df.loc[street_alcohol_df['Calle']==0]
street_alcohol_df.drop(index_where_zero.index)
```

```python
temp =street_alcohol_df.loc[street_alcohol_df['Por alcohol'] != 0 ].count()
temp['Por alcohol']
```

```python
prob = temp['Por alcohol'] / street_alcohol_df['Por alcohol'].count() 
prob
```

```python
garza = street_alcohol_df.loc[street_alcohol_df['Calle']=='GARZA SADA']
occ = garza.count().Calle
occ
```
